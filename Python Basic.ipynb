 ## **Q2.What is an interpreter in Python?**

An interpreter in Python is a program that executes Python code line by line rather than compiling the entire code before running it. It translates high-level Python code into machine-readable instructions at runtime.

## **Q5.What is Mutability in Python?**


In Python, mutability refers to whether an object's state (its data) can be changed after it is created.

* **Mutable Objects:** Can be modified after creation.
* **Immutable Objects:** Cannot be changed after creation.

**Mutable vs. Immutable Objects in Python**


Screenshot 2025-01-30 205333.png

**Example: Mutable (List)**

```python
my_list = [1, 2, 3]
my_list[0] = 99  # ✅ Allowed (List is mutable)
print(my_list)   # Output [99, 2, 3]
```

**Example: Immutable (String)**

```python
my_string = "hello"
my_string[0] = "H"  #  TypeError: 'str' object does not support item assignment
```

**Instead, you must create a new string:**

```python
new_string = "H" + my_string[1:]  # Correct way
print(new_string)  # "Hello"
```

**Why is Mutability Important?**

**1.Performance & Memory Efficiency**
* Immutable objects are hashable and used as dictionary keys.

* Mutable objects prevent accidental overwrites.

**2.Function Arguments & Side Effects**
* Mutable objects can change in-place inside functions.

* Immutable objects ensure no accidental modifications.

**Example: Mutable List as Function Argument**

```python
def modify_list(lst):
    lst.append(4)  # ✅ List is modified in place

my_list = [1, 2, 3]
modify_list(my_list)
print(my_list)  # Output: [1, 2, 3, 4] (Modified)
```

**Example: Immutable String as Function Argument**


```python
def modify_string(s):
    s += " world"  #  A new string is created (original remains unchanged)

my_str = "hello"
modify_string(my_str)
print(my_str)  # Output "hello" (Unchanged)
```

## **Q6.Why are lists mutable, but tuples are immutable**

The key difference between lists and tuples in Python lies in mutability:

* Lists are mutable → meaning their contents can be changed (modified, added, or removed).

* Tuples are immutable → meaning their contents cannot be changed after creation.

### **Reasons Behind This Difference:**

**1.Memory and Performance Optimization:**

* Tuples are stored in a way that makes them faster and more memory-efficient compared to lists.

* Since tuples cannot be changed, Python can store them in a more optimized manner.

**2.Hashability & Use in Dictionaries/Sets:**

* Tuples are hashable (if they contain only hashable elements), meaning they can be used as keys in dictionaries or elements in sets.

* Lists, being mutable, cannot be used as dictionary keys because their values can change, leading to inconsistency in hashing.

**3.Data Integrity & Safety:**

* Tuples ensure that their values remain constant, which is useful when you want to store fixed data (e.g., coordinates, database records).

* Lists are designed to be dynamic, allowing easy modifications.
**Example Demonstration:**

```python
# Lists are mutable
my_list = [1, 2, 3]
my_list.append(4)  # Adding an element
my_list[0] = 10    # Modifying an element
print(my_list)  # Output [10, 2, 3, 4]

# Tuples are immutable
my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # This will raise a TypeError
```

## **Q7.What is the difference between “==” and “is” operators in Python**

In Python, == and is are both comparison operators, but they serve different purposes:

**1. == (Equality Operator)**
* Checks if the values of two objects are the same.

* Does not check whether the objects are the same in memory

* It compares the contents of the objects.
**Example:**

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  #  True, because the values inside both lists are the same
print(a is b)  #  False, because they are different objects in memory
```

**2. is (Identity Operator)**
* Checks if two variables refer to the same object in memory.

* It does not compare values but rather the object identity.

**Example:**

```python
x = [4, 5, 6]
y = x  # y refers to the same object as x

print(x == y)  #  True, because values are the same
print(x is y)  #  True, because both refer to the same object in memory

z = [4, 5, 6]  # A new list with the same values as x

print(x == z)  #  True, because values are the same
print(x is z)  #  False, because they are different objects in memory
```

## **Q8.What are logical operators in Python?**

**Logical Operators in Python**

Logical operators in Python are used to combine conditional statements. They return Boolean values (True or False) based on the conditions given.

**Types of Logical Operators:**

Screenshot 2025-02-06 175735.png

**1. and Operator (Logical AND)**
* Returns True only if both conditions are True.
* Otherwise, it returns False.

**Example:**

```python
x = 10
y = 20

print(x > 5 and y > 15)  #  True (Both conditions are True)
print(x > 15 and y > 15) #  False (First condition is False)
```

**2. or Operator (Logical OR)**
* Returns True if at least one condition is True.

* Returns False only if both conditions are False.

**Example:**

```python
a = 5
b = 8

print(a > 10 or b > 6)  #  True (One condition is True)
print(a > 10 or b < 6)  #  False (Both conditions are False)
```

**3. not Operator (Logical NOT)**
Reverses the Boolean value:
* True → False
* False → True

**Example:**

```python
x = True
print(not x)  # False (Reverses True to False)

y = 10 > 5
print(not y)  # False (10 > 5 is True, so `not True` is False)
```

**Logical Operators in Conditional Statements**

```python
age = 18
country = "USA"

if age >= 18 and country == "USA":
    print("Eligible to vote")  # True, prints "Eligible to vote"
else:
    print("Not eligible")
```

## **Q9.What is type casting in Python?**

**Type Casting in Python**

Type casting (or type conversion) is the process of converting a variable from one data type to another in Python.

**Types of Type Casting**
Python provides two types of type casting:

**1.Implicit Type Casting (Automatic Conversion)**

* Python automatically converts a smaller data type into a larger data type to avoid data loss.

* No manual intervention is required.

**2.Explicit Type Casting (Manual Conversion)**

* The programmer manually converts one data type to another using built-in functions.

* This is required when Python does not convert types automatically.

**1. Implicit Type Casting (Automatic)**

Python automatically converts smaller data types into larger ones (e.g., int → float).

**Example:**

```python
a = 5   # Integer
b = 2.5 # Float

result = a + b  # Python automatically converts 'a' (int) to float

print(result)   #  Output: 7.5
print(type(result))  #  Output: <class 'float'>
```

**2. Explicit Type Casting (Manual)**

```python
# Converting float to int
num = 10.8
new_num = int(num)  # Explicit conversion

print(new_num)  # ✅ Output: 10 (decimal part removed)
print(type(new_num))  # ✅ Output: <class 'int'>
```

**Example: String to Integer Conversion**

```python
s = "123"
num = int(s)  # Convert string to integer

print(num + 10)  #  Output: 133
print(type(num))  # Output: <class 'int'>
```

**Important:**
*If you try to convert "abc" to int, Python will throw an error:*

```python
int("abc")  #  ValueError: invalid literal for int()
```

## **Q12.How does the elif statement work?**


The elif (short for "else if") statement allows checking multiple conditions in a structured way. It is used after an if statement and before an else statement.

The elif statement works by:

**1.Checking the first if condition:**
* If True, the block executes, and the rest are skipped.
* If False, it moves to the next elif condition.

**2.Checking multiple elif conditions in order:**
* If an elif condition is True, its block executes, and the rest are skipped.
* If False, it moves to the next elif or else block.

**3.Executing the else block if none of the if or elif conditions are True.**

**Syntax of** elif



```python
if condition1:
    # Code to execute if condition1 is True
elif condition2:
    # Code to execute if condition2 is True
elif condition3:
    # Code to execute if condition3 is True
else:
    # Code to execute if none of the above conditions are True
```

**Example: Grading System Using** elif

```python
marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")  # Output: Grade: B
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: F")
```

**How It Works?**

* marks >= 90 → False (skips if block).

* marks >= 75 → True (executes elif block and skips the rest).

**Example: Age-Based Ticket Pricing**

```python
age = 10

if age < 5:
    print("Free Entry")
elif age < 18:
    print("Child Ticket - $10")  # Output: Child Ticket - $10
elif age < 60:
    print("Adult Ticket - $20")
else:
    print("Senior Ticket - $15")
```

**How It Works?**

* age < 5 → False (skips if block).
* age < 18 → True (executes elif block and skips the rest).

```python
password = ""

while password != "1234":
    password = input("Enter password: ")

print("Access granted!")
```

## **Q14.Describe a scenario where a while loop is more suitable than a for loop.**

A while loop is more suitable when the number of iterations is unknown and depends on a condition being met.

**Scenario: User Input Validation**

Imagine you are building a login system where a user must enter the correct password to gain access. Since you don't know how many attempts the user will take, a while loop is the best choice.

**Example: Login System Using while Loop**

```python
correct_password = "secret123"
user_input = ""

while user_input != correct_password:
    user_input = input("Enter your password: ")

print("Access granted!")
```

**Why Use a while Loop?**

* The loop runs until the user enters the correct password.

* The number of attempts is unknown in advance.

* A for loop would not work well here because we don’t have a fixed number of iterations.


**Other Real-World Scenarios for while Loops**

**1.Waiting for a condition to be met** (e.g., waiting for a file to be downloaded).

**2.Game loops** (e.g., running a game until the player quits).

**3.ATM withdrawal limit enforcement** (keep asking for the correct PIN).

**4.Processing live data** (e.g., reading sensor data continuously).

## **Q13.Write a program to reverse a string using a while loop.**

```python
# Get user input
string = input("Enter a string: ")

# Initialize variables
reversed_string = ""
index = len(string) -1 # start from the last index

# Loop through the string in reverse order
while index >= 0:
  reversed_string += string[index] # Append each character in reverse order
  index -= 1 # Move to the previous character

# Print the reversed string
print("Reversed string:", reversed_string)
```

## **Q14.Write a Python program to calculate the factorial of a number provided by the user using a while loop.**


```python
# Get user input
num = int(input("Enter a number:"))

# Intialize factorial variable
factorial = 1

# Initialize counter
i = num

# Calculate factorial using while loop
while i > 0:
  factorial *= i # Multiply current number
  i -= 1 # Decrease counter

# Print the factorial
print(f"Factorial of {num} is {factorial}")
```
